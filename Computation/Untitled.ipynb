```python
def logistic(r, x0, step, datatype):
    x = np.zeros(step, dtype=datatype)
    x[0]= x0
    for i in range(step-1):
        x[i+1] = r*x[i]*(1-x[i])
    return x
```

```python
y64 = logistic(3.9, 0.4, 50, np.float64)
y32 = logistic(3.9, 0.4, 50, np.float32)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(1,50,50)
plt.plot(x,y32,'r-')
plt.plot(x,y64)
```
